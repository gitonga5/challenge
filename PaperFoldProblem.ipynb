##### 【Problem 1 】 Implementation using the power arithmetic operator

```python
#Code to calculate the thickness of a piece of paper when folded 43 times

THICKNESS = 0.00008
n_folds = 43

folded_thickness = THICKNESS * (2**n_folds)

print("Thickness: {} Meters".format(folded_thickness))
```

##### 【Problem 2 】 Unit conversion



```python
#Convert meters to kilometers and display with two decimal places.
print("Thickness： {:.2f} Kilometers".format(folded_thickness / 1000))

#Convert meters to Million kilometers and display with two decimal places.
print("Thickness： {:.2f} Million Kilometers".format(folded_thickness / 1000000))

```

```python
# use for loops
THICKNESS = 0.00008
n_folds = 43

folded_thickness = THICKNESS

for i in range(n_folds):
    folded_thickness = folded_thickness * 2

print("Thickness: {} Meters".format(folded_thickness))

thickness_km = folded_thickness / 1000
print("Thickness: {:.2f} Kilometers".format(thickness_km))

thickness_million_km = folded_thickness / 1000000
print("Thickness: {:.2f} Million Kilometers".format(thickness_million_km))

```

##### 【Problem 5 】 Save to list

```python

THICKNESS = 0.00008
FOLD_COUNT = 43

thickness_list = []

thickness_list.append(THICKNESS)

folded_thickness = THICKNESS

for _ in range(FOLD_COUNT):
    folded_thickness *= 2
    thickness_list.append(folded_thickness)
```

##### 【Problem 6 】 Display line chart

```python
# To draw a graph Matplotlib Is library Use Use the following chick shape behind the code you want to record to the list.

import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Thickness of Folded Paper")
plt.xlabel("Number of Folds")
plt.ylabel("Thickness [m]")
plt.plot(thickness_list)
plt.show()


#in the early folds, the thickness increases slowly but there ia a massive increase from the 40th fold.

```
